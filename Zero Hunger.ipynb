```python
yield_per_year = yield_data.groupby('Year').sum()
yield_per_year.Total
```

```python
yield_per_crop = yield_data.groupby('Crop').sum()
yield_per_crop.Total
```

```python
crops = ['Bengal Gram','Groundnut','Maize']
plt.pie(yield_per_crop.Total,labels=crops)
plt.title('Overall Yield per crop')
plt.plot()
```

- Inferences:
1.Among Maize, Groundnut and Bengal Gram, Maize has the highest Seasonal yield.

2.Nagarkurnool and Mahabubnagar are the only districts where yield of Groundnut is higher than that of Maize.

3.Hyderbad has no yield.

4.Adilabad has the highest yield of all the crops combined.

5.2017-2018 has the highest overall yield.

6.The production of Rice is very high compared to production of Maize.

7.Nizamabad's production is much higher than the production in Nagarkurnool and Karimnagar.

8.Most of the blocks in different districts have increasingly high production in 2019 except for Gangadhara, Gannervaram, Karimnagar_Rural(significantly less than previous year), Timmapur, Balkonda, Kotgiri, Makloor, Navipet, Vami.

9.Blocks like Biginapalle,Kodair, peddakothapalle, Telkapalle have show significant increase in the annual yield from 2018 to 2019.

10.The least combined yield was in the year 2016 and highest in 2019 for rice and maize during kharif season.

11.During rabi season bhodan has the highest yield compared to all others followed by kotgiri.

12.Groundnut has significant increase in the yield over years.

13.Medchal Malkajgiri had no contribution towards the yield in groundnut.

14.Warnagal Rural had the district wise highest yield in both Groundnut and Maize where as Wanaparthy had highest yield in Bengal gram.
